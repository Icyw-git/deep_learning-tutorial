```python
#图像相关知识
#图像是由像素组成的二维矩阵,
# 每个像素都有一个颜色值,
# 颜色值可以表示为RGB值,
# 即红色、绿色、蓝色的组合。
# 每个像素的颜色值可以用一个3维向量表示,
# 例如(255,0,0)表示红色,
# (0,255,0)表示绿色,
# (0,0,255)表示蓝色。
# 图像可以表示为一个3维矩阵,
# 第一个维度表示颜色通道,
# 第二个维度表示高度,
# 第三个维度表示宽度。
# 例如,一个28x28的灰度图像可以表示为一个2维矩阵,
# 每个元素表示一个像素的灰度值。
# 而一个28x28的彩色图像可以表示为一个3维矩阵,
# 第一个维度表示红色通道,
# 第二个维度表示绿色通道,
# 第三个维度表示蓝色通道。
# 每个元素表示一个像素的颜色值。

#图像的保存和绘制和读取
#使用matlib.pyplot库可以方便地保存和绘制图像。
# 例如,可以使用plt.imshow()函数绘制图像,
# 使用plt.savefig()函数保存图像。
# 可以使用plt.imread()函数读取图像。
# 例如,可以使用plt.imread('image.jpg')读取名为image.jpg的图像。
# 读取的图像将返回一个3维矩阵,
# 第一个维度表示高度,
# 第二个维度表示宽度,
# 第三个维度表示颜色通道。

#示例
import matplotlib.pyplot as plt
img=plt.imread('photo.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
print(img.shape)
print(img)
# plt.imsave('王楚之_gray.jpg',img,cmap='gray')
# 可以使用plt.imsave()函数将图像保存为灰度图像。
# 例如,可以使用plt.imsave('王楚之_gray.jpg',img,cmap='gray')将王楚之.jpg保存为灰度图像。
```

```python
#CNN卷积模型
#由卷积层、池化层、全连接层等组成。
#卷积层：用于提取图像的特征,
# 通过卷积核在图像上滑动,
# 对每个位置的像素进行点积操作,
# 生成特征图。
#池化层：用于减少特征图的空间维度,
# 保留重要特征,
# 防止过拟合。
#全连接层：用于将特征图映射到输出类别。

#最开始是数据输入层,对数据进行预处理,
# 例如,归一化、标准化、resize等操作,
# 以确保输入数据的质量和一致性。
# 用于接收输入的图像数据。
# 图像数据通常是一个3维矩阵,
# 第一个维度表示高度,
# 第二个维度表示宽度,
# 第三个维度表示颜色通道。

#中间是卷积层,
# 用于提取图像的特征,
# 通过卷积核在图像上滑动,
# 对每个位置的像素进行点积操作,
# 生成特征图。
# 卷积层通常包含多个卷积核,
# 每个卷积核都有不同的特征提取能力,
# 从而提取出不同的特征。

# 每个卷积核的参数数量是其大小（高度*宽度）加上一个偏置项,
# 例如,一个3x3的卷积核有9个参数（3x3=9）加上一个偏置项=10个参数。
# 卷积层的输出是一个4维矩阵,
# 第一个维度表示样本数量,
# 第二个维度表示卷积核数量,
# 第三个维度表示特征图的高度,
# 第四个维度表示特征图的宽度。

#然后是激励层,
# 用于引入非线性变换,
# 使模型能够学习复杂的特征表示。
# 常用的激励层包括ReLU、Sigmoid、Tanh等。
# ReLU（Rectified Linear Unit）是一种常用的激励层,
# 它将所有负数设为0,
# 而将所有正数保持不变。
# ReLU的计算简单,
# 并且在深度神经网络中被广泛使用,
# 因为它能够加速模型的收敛并提高模型的性能。

#之后是池化层,
# 用于减少特征图的空间维度,
# 保留重要特征,
# 防止过拟合。
# 常用的池化层包括最大池化层和平均池化层。
# 最大池化层在每个池化窗口中选择最大值作为输出,
# 而平均池化层则计算池化窗口中所有像素的平均值作为输出。
# 池化层的输出是一个2维矩阵,
# 第一个维度表示样本数量,
# 第二个维度表示特征图的高度和宽度（由于池化层减少了空间维度,
# 所以高度和宽度会减半）。

#最后是全连接层,
# 用于将特征图映射到输出类别。
# 全连接层通常包含多个神经元,
# 每个神经元对应一个输出类别。
# 全连接层的输出是一个1维向量,
# 每个元素表示对应类别的得分（或概率）。
# 为了将得分转换为概率,
# 通常会使用softmax函数对输出进行归一化。
# 例如,如果有3个输出类别,
# 则全连接层的输出可能是[0.2, 0.5, 0.3],
# 表示该样本属于第一个类别、第二个类别和第三个类别的概率分别为20%、50%和30%。

```

```python
#卷积计算的规则
#1. 卷积核在图像上滑动,
# 对每个位置的像素进行点积操作,
# 生成特征图的一个像素。
#2. 卷积核的大小通常是一个小的矩形,
# 例如3x3或5x5,
# 它在图像上滑动时,
# 每个位置的像素与卷积核进行点积操作,
# 生成一个特征图的像素。
#3. 卷积核在图像上滑动时,
# 每个位置的像素都要与卷积核进行点积操作,
# 因此,
# 卷积计算的总次数是图像大小与卷积核大小的乘积,
# 例如,
# 一个32x32的图像与3x3的卷积核进行卷积计算,
# 总次数是32x32x9=9216次。

#卷积层的主要作用：
#1.特征提取：
# 卷积层通过卷积核在图像上滑动,
# 提取图像的局部特征,
# 例如,
# 边缘、纹理、形状等。
#2.参数共享：
# 卷积层中的每个卷积核参数在图像上滑动时都是共享的,
# 这意味着无论图像有多大,
# 卷积核的参数数量都是固定的,
# 这使得卷积层在处理大规模图像时具有高效性。
#3.平移不变性：
# 卷积层对图像的平移不变性很强,
# 这意味着无论图像在空间上如何平移,
# 卷积核提取的特征都是相同的,
# 这使得卷积层在处理图像分类等任务时具有优势。
#4.局部连接：
# 卷积层中的每个卷积核只与图像中的局部区域进行连接,
# 这使得卷积层在处理大规模图像时具有高效性,
# 同时也减少了参数数量,
# 防止过拟合。

#局部感受野：
# 卷积层中的每个卷积核只与图像中的局部区域进行连接,
# 这使得卷积层在处理大规模图像时具有高效性,
# 同时也减少了参数数量,
# 防止过拟合。
# 例如,
# 一个32x32的图像与3x3的卷积核进行卷积计算,
# 每个卷积核只与图像中的9个像素进行连接,
# 总参数数量是3x3x1=9个,
# 而不是32x32x9=9216个。
#一个卷积核就是一个神经元,
# 它的输出就是特征图的一个像素。
```

```python
#Padding填充
#Padding是指在图像的边界上添加额外的像素层,
# 以保持图像的尺寸不变。
# 这在卷积操作中非常重要,
# 因为卷积操作会导致图像的尺寸减小,
# 而Padding可以确保卷积操作在边界上的像素也能被处理。
# 常用的Padding方法有零填充（Zero Padding）和反射填充（Reflection Padding）。
# 零填充是在图像边界上添加全零像素,
# 而反射填充是在图像边界上镜像反射像素。
# 这两种方法都可以保持图像的尺寸不变,
# 同时允许卷积操作在边界上的像素也能被处理。

#作用
#1. 保持图像尺寸不变：Padding可以确保卷积操作在边界上的像素也能被处理,
# 从而保持图像的尺寸不变。
#2. 防止信息丢失：Padding可以防止卷积操作在边界上丢失重要信息,
# 从而提高模型的性能。
#3. 提高模型的鲁棒性：Padding可以提高模型的鲁棒性,
# 使其对输入数据的变化更加敏感。

#padding的类型以及选择
#1.Valid Padding
# 不添加任何填充层,
# 卷积操作在图像边界上的像素也不会被处理。
# 这意味着输出特征图的尺寸会比输入图像小,
# 从而丢失了图像边界上的信息。

# 2. Same Padding
# 填充层的数量足够,
# 使得卷积操作在图像边界上的像素也能被处理。
# 这意味着输出特征图的尺寸与输入图像相同,
# 从而保留了图像边界上的信息。


#Stride步长
#Stride是指卷积核在图像上滑动的步长,
# 它决定了卷积操作在图像上的采样率。
# 通常,
# Stride的默认值为1,
# 这意味着卷积核在图像上每次滑动一个像素。
# 但是,
# 当Stride大于1时,
# 卷积核在图像上每次滑动多个像素,
# 从而减少了输出特征图的尺寸。
# 这在某些情况下可以减少计算量,
# 同时也可以提取更高级的特征。

#作用
#1. 减少计算量：Stride大于1时,
# 卷积核在图像上每次滑动多个像素,
# 从而减少了输出特征图的尺寸,
# 从而减少了计算量。
#2. 提取更高级的特征：Stride大于1时,
# 卷积核在图像上每次滑动多个像素,
# 从而提取了更高级的特征,
# 这在某些任务中是非常重要的。
#3.增大感受野：Stride大于1时,
# 卷积核在图像上每次滑动多个像素,
# 从而增加了感受野,
# 这使得模型能够提取更广泛的特征。
```

```python
#多通道卷积计算
#多通道卷积计算与单通道卷积计算类似,
# 只是在输入图像和卷积核上都有多个通道。
# 每个通道的卷积操作独立进行,
# 最后将所有通道的输出特征图拼接起来。
# 例如,
# 如果输入图像有3个通道,
# 卷积核也有3个通道,
# 则每个通道的卷积操作独立进行,
# 最后将3个通道的输出特征图拼接起来,按位相加
# 得到一个多通道的输出特征图。

#多卷积核卷积计算
#多卷积核卷积计算与多通道卷积计算类似,
# 只是在卷积核上有多个卷积核。
# 每个卷积核的卷积操作独立进行,
# 最后将所有卷积核的输出特征图拼接起来。
# 例如,
# 如果卷积核有3个卷积核,
# 则每个卷积核的卷积操作独立进行,
# 最后将3个卷积核的输出特征图拼接起来,按位相加
# 得到一个多卷积核的输出特征图，有三张特征图

#特征图大小计算
#特征图的大小可以通过以下公式计算：
# 特征图的大小=（输入图像的大小-卷积核的大小+2*填充）/步长  +1
# 例如,
# 如果输入图像的大小为28x28,
# 卷积核的大小为3x3,
# 填充为1,
# 步长为1,
# 则输出特征图的大小为28x28。
# 如果填充为2,
# 则输出特征图的大小为30x30。
```

```python
#卷积层的api介绍
#torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros')
#参数：
#in_channels：输入图像的通道数
#out_channels：输出图像的通道数
#kernel_size：卷积核的大小
#stride：卷积核的步长
#padding：填充层的数量
#dilation：卷积核的膨胀率
#groups：分组卷积的组数
#bias：是否添加偏置项
#padding_mode：填充模式，默认值为'zeros'

#示例
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
img=plt.imread('photo.jpg') #数据导入
print(img.shape)
input=torch.tensor(img,dtype=torch.float32) #将数据转换为张量
input=input.permute(2,0,1) #将通道维放到第一维
input=input.unsqueeze(0) #在第一维添加一个样本维
print(input.shape)

cnn=nn.Conv2d(in_channels=3,out_channels=5,kernel_size=3,stride=1,padding=1) #卷积层
output=cnn(input) #前向传播
print(output) #输出张量的形状为(1,5,512,512)
print(output.shape)

#可视化
plt.imshow(output[0,2,:,:].detach().numpy()) #可视化第五个通道的输出
plt.title('Conv2d')
plt.show()


```

```python
#池化层
#池化层的作用是减少特征图的空间维度,
# 从而减少参数数量和计算量,
# 同时保留重要的特征信息。
# 池化层通常用于卷积层之后,
# 用于下采样和特征提取。
# 池化层的种类有最大池化层和平均池化层。
# 最大池化层取池化窗口中的最大值作为输出,
# 而平均池化层取池化窗口中的平均值作为输出。

#池化层与padding
#池化层与padding的关系是,
# 池化层通常会与padding层一起使用,
# 以保持特征图的空间维度不变。
# 例如,
# 最大池化层通常会与padding层一起使用,
# 以确保池化窗口不会超出特征图的边界。

#多通道池化
#多通道池化层的作用是对每个通道独立进行池化操作,
# 从而保留每个通道的特征信息。
# 多通道池化层通常用于多通道特征图的池化操作,
# 例如,
# 对每个通道的特征图进行最大池化操作,
# 以提取每个通道的重要特征。

#池化层的api介绍
#torch.nn.MaxPool2d(kernel_size, stride=None, padding=0, dilation=1, return_indices=False, ceil_mode=False)
#参数：
#kernel_size：池化窗口的大小
#stride：池化窗口的步长
#padding：填充层的数量
#dilation：池化窗口的膨胀率
#return_indices：是否返回池化窗口的索引
#ceil_mode：是否使用向上取整的模式
#torch.nn.AvgPool2d(kernel_size, stride=None, padding=0, ceil_mode=False, count_include_pad=True)
#参数：
#kernel_size：池化窗口的大小
#stride：池化窗口的步长
#padding：填充层的数量
#ceil_mode：是否使用向上取整的模式
#count_include_pad：是否包括填充层的像素
```

```python
#示例
input=plt.imread('photo.jpg')
plt.imshow(input)
plt.show()
print(input.shape)
input=torch.from_numpy(input).float() #将numpy数组转换为torch张量，数据类型为float32
input=input.permute(2,0,1)
input.unsqueeze_(0)   #下划线表示就地修改
print(input.shape)

cnn=nn.Conv2d(in_channels=3,out_channels=4,kernel_size=3,stride=1,padding=1)
output=cnn(input)
print(output.shape)

pool=nn.MaxPool2d(2,1) #池化层，2*2的池化窗口，步长为1
output=pool(output)
print(output.shape)
#可视化
plt.imshow(output[0,3].detach().numpy())  #可视化第0个样本的第3个通道
plt.show()
```
