```python
#RNN神经网络介绍
#RNN神经网络是一种循环神经网络，它可以处理序列数据，例如文本、语音、视频等。
#RNN神经网络的基本思想是将输入序列中的每个元素都映射到一个固定长度的向量表示，然后使用这个向量表示来预测下一个元素。
#RNN神经网络的一个重要问题是梯度消失问题，即当序列长度增加时，梯度会变得非常小，导致模型无法学习长距离依赖关系。
#为了解决这个问题，提出了许多改进的RNN模型，例如LSTM模型、GRU模型等。

#自然语言处理
#自然语言处理是指将人类语言转换为计算机可以理解和处理的形式。
#自然语言处理的一个重要任务是文本分类，例如情感分析、垃圾邮件过滤等。
#为了处理文本数据，提出了许多基于RNN神经网络的模型，例如RNN模型、LSTM模型、GRU模型等。
#NLP模型
#NLP模型是指基于NLP任务的模型，例如文本分类、情感分析、命名实体识别等。
#NLP模型的基本思想是将文本转换为向量表示，然后使用RNN神经网络来处理这个向量表示。
#为了提高NLP模型的性能，提出了许多改进的模型，例如BERT模型、GPT模型等。
#模型的结构
#NLP模型的结构通常包括以下几个部分：
#1. 嵌入层（Embedding Layer）：将文本转换为向量表示。
#2. RNN层（RNN Layer）：处理向量表示的序列数据。
#3. 全连接层（Fully Connected Layer）：将RNN层的输出映射到最终的分类结果。
```

```python
#词嵌入层
#词嵌入层是将文本中的每个单词转换为一个固定长度的向量表示。
#词嵌入层的基本思想是将单词映射到一个低维空间，使得相似的单词在这个空间中距离更近。
#为了实现词嵌入层，提出了许多模型，例如Word2Vec模型、GloVe模型等。
#比如有100个不同的单词，每个单词用一个10维向量表示，那么词嵌入层的输出就是一个100x10的矩阵。

#词嵌入层的工作流程
#1. 输入：一个文本序列，例如一个句子或一个段落。
#2. 预处理：将文本序列中的每个单词转换为一个整数索引，例如将"hello"转换为100，"world"转换为200等。
#3. 词嵌入：使用预训练的词嵌入模型，将每个整数索引转换为一个固定长度的向量表示。
#4. 输出：一个矩阵，其中每一行表示一个单词的向量表示。
#例如，对于句子"hello world"，假设"hello"的索引为100，"world"的索引为200，那么词嵌入层的输出就是一个2x10的矩阵，其中第一行是"hello"的向量表示，第二行是"world"的向量表示。

#词嵌入层的使用
#1. 初始化：在模型中添加一个词嵌入层，指定词汇表的大小和向量的维度。
#2. 前向传播：将输入的整数索引序列传递给词嵌入层，得到对应的向量表示。
#3. 后续层：将词嵌入层的输出传递给后续的层，例如循环神经网络（RNN）或卷积神经网络（CNN）。
#api
#torch.nn.Embedding(num_embeddings, embedding_dim, padding_idx=None, max_norm=None, norm_type=2.0, scale_grad_by_freq=False, sparse=False, _weight=None)
#num_embeddings：词汇表的大小
#embedding_dim：向量的维度
#padding_idx：如果指定了padding_idx，那么在计算损失时会忽略该索引对应的向量表示。
#max_norm：如果指定了max_norm，那么每个向量的范数不会超过max_norm。
#norm_type：范数的类型，默认是2范数。
#scale_grad_by_freq：如果设置为True，那么每个向量的梯度会根据其在训练数据中出现的频率进行缩放。
#sparse：如果设置为True，那么词嵌入层的权重矩阵会被稀疏化，从而减少内存占用。
```

```python
#词嵌入层的演示
#导包
import torch
import torch.nn as nn
import jieba
```

```python
text='北京冬奥的进度条已经过半，不少外国运动员在完成自己的比赛之后踏上归途。'
word=jieba.lcut(text)
print(word)
embed=nn.Embedding(num_embeddings=len(word),embedding_dim=4)

for i,word in enumerate(word):  #enumerate作用是将一个可遍历的数据对象(如列表、元组或字符串)组合为一个索引序列，同时列出数据和数据下标
    print(i,word)
    word_vector=embed(torch.tensor(i))  #将i转换为张量，因为embed层的输入必须是张量,然后通过embed层获取对应的词向量
    print(f'word vector of {word}:{word_vector}') #打印每个单词的词向量
```

```python
#循环网络层
#循环网络层是一种特殊的神经网络层，用于处理序列数据。
#循环网络层的基本思想是将序列中的每个元素依次输入到网络中，每个元素的输入都依赖于之前元素的输出。
#为了实现循环网络层，提出了许多模型，例如RNN模型、LSTM模型、GRU模型等。

#RNN的计算过程
#假设输入序列为x=(x1,x2,x3,...,xt)，其中xt是第t个时间步的输入。
#假设隐藏状态为h=(h1,h2,h3,...,ht)，其中ht是第t个时间步的隐藏状态。
#假设输出序列为y=(y1,y2,y3,...,yt)，其中yt是第t个时间步的输出。
#假设RNN模型的参数为Wih、Whh、Why和bh、bh和by，其中Wih是输入到隐藏状态的权重矩阵，Whh是隐藏状态到隐藏状态的权重矩阵，Why是隐藏状态到输出的权重矩阵，bh是隐藏状态的偏置向量，by是输出的偏置向量。
#则RNN模型的计算过程如下：
#ht=σ(Wihxt+Whhht−1+bh)
#yt=Whyht+by
#其中，σ是激活函数，例如sigmoid函数或tanh函数。
#ht是第t个时间步的隐藏状态，依赖于第t-1个时间步的隐藏状态和第t个时间步的输入。
#yt是第t个时间步的输出，依赖于第t个时间步的隐藏状态。

#词汇表映射
#在RNN模型中，输入序列中的每个元素都是一个整数，称为词汇表索引。
#为了将词汇表索引转换为向量表示，需要使用词汇表映射层。
#词汇表映射层是一个简单的全连接层，将词汇表索引映射为向量表示。
#假设词汇表大小为V，向量表示维度为d，则词汇表映射层的参数为Wv，其中Wv是一个V行d列的矩阵，第i行表示第i个词汇表索引的向量表示。
#则词汇表映射层的计算过程如下：
#x̂t=Wvxt
#其中，x̂t是第t个时间步的输入向量表示，xt是第t个时间步的词汇表索引。

#文本生成示例
#假设我们有一个文本数据集，其中每个样本是一个字符串。
#我们的目标是使用RNN模型生成新的文本。
#为了实现这一目标，我们需要将文本中的每个字符转换为一个整数，称为字符索引。
#然后，我们可以使用RNN模型生成新的字符索引序列，将其转换为字符串，即可生成新的文本。
#例如，假设我们的RNN模型生成了一个字符索引序列为[10,15,20,25,30]，则可以将其转换为字符串为"hello"。
```

```python
#rnn层 api
#rnn=nn.RNN(input_size,hidden_size,num_layers,batch_first=True)
#input_size:输入特征的维度，一般为词嵌入层的输出维度
#hidden_size:隐藏层的维度
#num_layers:rnn层的层数
#batch_first:如果为True,则输入和输出张量的形状为(batch_size,seq_len,input_size)
#否则为(seq_len,batch_size,input_size)

#api演示
x=torch.rand(size=(5,32,128)) #各维度分别为序列长度、批次大小和输入特征维度，序列长度意思是每个样本的时间步数，批次大小是每次输入模型的样本数量，输入特征维度是每个时间步输入的特征数量,也是词向量维数
h0=torch.rand(size=(1,32,256)) #各维度分别为层数、批次大小和隐藏层维度，层数是rnn层的层数，批次大小是每次输入模型的样本数量，隐藏层维度是每个时间步输出的特征数量
rnn=nn.RNN(input_size=128,hidden_size=256,num_layers=1) #输入特征维度为128，隐藏层维度为256，rnn层的层数为1
x1,h1=rnn(x,h0)
print(x1.shape)
print(h1.shape)

```
