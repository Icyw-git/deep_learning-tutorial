```python
#反向传播
#反向传播是指在神经网络中，从输出层向输入层方向，计算每个神经元对最终损失的贡献度，从而更新每个神经元的权重和偏置。
#反向传播的过程如下：
#1. 计算输出层的误差项，即预测值与真实值之间的差异。
#2. 从输出层向隐藏层方向，计算每个隐藏层神经元的误差项，即该神经元对最终损失的贡献度。
#3. 更新每个隐藏层神经元的权重和偏置，使损失函数的值变小。
#4. 重复步骤2和3，直到达到预设的迭代次数或损失函数的值足够小。

#指数移动平均加权
#指数移动平均加权是指在计算移动平均线时，给最近的数据赋予更高的权重， older data赋予较低的权重。
#移动平均线的计算公式如下：
#MA(t)=α⋅X(t)+(1−α)⋅MA(t−1)
#其中，MA(t)表示第t个时间点的移动平均线，X(t)表示第t个时间点的数据，α表示权重系数，MA(t−1)表示第t−1个时间点的移动平均线。
#当α=1时，移动平均线等于当前数据点，即不考虑历史数据。
#当α<1时，移动平均线对历史数据的权重较低，对当前数据的权重较高。
#移动平均线可以用于平滑数据，减少噪声和异常值的影响。

#梯度下降优化方法：动量法
#动量法是指在梯度下降优化过程中，引入动量项，使优化过程更加稳定和快速。
#梯度计算公式：s_t=b*s_{t-1}+(1-b)*g_t
#参数更新公式：w_t=w_{t-1}-η*s_t
#其中，s_t表示第t个时间点的动量项，b表示动量系数，s_{t-1}表示第t−1个时间点的动量项，g_t表示第t个时间点的梯度。
#当b=0时，动量法等于普通的梯度下降优化方法。
#当b<1时，动量法对历史梯度的权重较低，对当前梯度的权重较高。
#动量法可以用于处理局部最优解问题，使优化过程更加稳定。
#动量法的作用是：
#1. 加速优化过程：动量项可以使优化过程更加快速，跳过局部最优解，直接到达全局最优解。
#2. 提高优化稳定性：动量项可以平滑梯度更新，减少震荡和波动，提高优化稳定性。

#api:
#torch.optim.SGD(params, lr=0.01, momentum=0, dampening=0, weight_decay=0, nesterov=False)
#参数：
#params：待优化的参数组
#lr：学习率，默认值为0.01
#momentum：动量系数，默认值为0
#dampening：动量衰减系数，默认值为0
#weight_decay：权重衰减系数，默认值为0
#nesterov：是否使用Nesterov动量，默认值为False
#示例
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from torch import nn

w=torch.tensor([2],dtype=torch.float32,requires_grad=True) #requires_grad=True表示需要计算梯度
w_list=[]
for epoch in range(100):
    loss=w**2
    optimizer=torch.optim.SGD(params=[w],lr=0.1,momentum=0.9)#params表示需要优化的参数,lr表示学习率,momentum表示动量系数
    optimizer.zero_grad() #将梯度清零,否则会累加
    loss.sum().backward()
    optimizer.step() #更新参数
    print(w)
    w_list.append(w.item())
plt.figure(figsize=(10,6))
sns.lineplot(x=range(100),y=w_list)
plt.title('Momentum Optimization')
plt.xlabel('Epoch')
plt.ylabel('w')
plt.show()

```

```python
#梯度下降优化方法——adagrad
#adagrad是一种自适应学习率的优化方法,
# 它根据每个参数的历史梯度信息来调整学习率,
# 使得在训练过程中对不同参数的更新有不同的学习率,
# 从而加速收敛并提高模型的性能。

#计算步骤
#1.初始化学习率和参数，小常数ε=1e-8
#2.初始化梯度累计变量s=0
#3.从训练集里面采样，计算梯度g_t
#4.累计梯度平方和：s_t=s_t-1+g_t^2
#5.学习率计算公式：α/（sqrt（s_t）+ε）
#6.更新参数：w_t=w_t-1-α*g_t/（sqrt（s_t）+ε）

#方法的优缺点
#1. 优点：
# （1）自适应学习率：adagrad根据每个参数的历史梯度信息来调整学习率,
# 使得在训练过程中对不同参数的更新有不同的学习率,
# 从而加速收敛并提高模型的性能。
# （2）简单实现：adagrad的计算步骤简单,
# 只需要维护一个梯度累计变量和一个学习率即可,
# 实现起来比较方便。
#2. 缺点：
# （1）学习率_decay：adagrad的学习率_decay参数比较敏感,
# 如果设置得过小,
# 会导致学习率_decay过慢,
# 从而影响模型的收敛速度;
# 如果设置得过大,
# 会导致学习率_decay过快,
# 从而影响模型的性能。
# （2）内存占用：adagrad需要维护一个梯度累计变量,
# 内存占用较大,
# 特别是在处理大规模数据集时,
# 会占用较多的内存空间。

#api
#torch.optim.Adagrad(params, lr=0.01, lr_decay=0, weight_decay=0, initial_accumulator_value=0, eps=1e-10)
#参数：
#params：待优化的参数组
#lr：学习率，默认值为0.01
#lr_decay：学习率_decay系数，默认值为0
#weight_decay：权重衰减系数，默认值为0
#initial_accumulator_value：初始梯度累计变量值，默认值为0
#eps：小常数，默认值为1e-10

#示例
w=torch.tensor([4],dtype=torch.float32,requires_grad=True)

optimizer=torch.optim.Adagrad(params=[w],lr=0.1,eps=1e-8)
w_list=[]
for epoch in range(100):
    loss=0.5*w**2+10
    optimizer.zero_grad()
    loss.sum().backward()
    optimizer.step()
    w_list.append(w.item())
    print(w)
plt.figure(figsize=(10,6))
sns.lineplot(x=range(100),y=w_list)
plt.title('Adagrad Optimization')
plt.xlabel('Epoch')
plt.ylabel('w')
plt.show()
```

```python
#梯度优化方法——RMSProp
#RMSProp是一种基于梯度的优化方法,是对Adagrad的改进,
# 它通过对梯度的平方进行指数加权移动平均,
# 来调整学习率,
# 从而加速收敛并避免局部最优解。
#计算方法
#1.初始化学习率，权重参数，小常数ε
#2.初始化梯度累计变量s=0
#3.从训练集中随机抽取多个样本，计算梯度g_t
#4.使用指数加权移动平均更新s_t：s_t=γs_{t-1}+(1-γ)g_t^2
#5.使用学习率调整权重参数：w_t=w_{t-1}-η/sqrt(s_t+\varepsilon) *g_t

#优缺点
#1. 优点：
# （1）自适应学习率：RMSProp根据梯度的平方进行自适应调整学习率,
# 可以在不同维度上有不同的学习率,
# 从而加速收敛。
# （2）平滑梯度：RMSProp通过对梯度的平方进行指数加权移动平均,
# 可以平滑梯度更新,
# 避免震荡和波动。
# （3）处理稀疏梯度：RMSProp对稀疏梯度有较好的处理能力,
# 可以在处理大规模数据集时保持较好的性能。
#2. 缺点：
# （1）学习率衰减问题：RMSProp的学习率衰减是指数级的,
# 可能会导致学习率过早衰减,
# 从而影响收敛速度。
# （2）对初始学习率敏感：RMSProp对初始学习率敏感,
# 不同的初始学习率可能会导致不同的收敛结果。

#示例
w=torch.tensor([4],dtype=torch.float,requires_grad=True)
optimizer=torch.optim.RMSprop(params=[w],lr=0.1,eps=1e-8)
w_list=[]
for epoch in range(100):
    loss=0.5*w**2+20
    optimizer.zero_grad()
    loss.sum().backward()
    optimizer.step()
    w_list.append(w.item())
    print(w)
plt.figure(figsize=(10,6))
sns.lineplot(x=range(100),y=w_list)
plt.title('RMSProp Optimization')
plt.xlabel('Epoch')
plt.ylabel('w')
plt.show()
```

```python
#梯度下降优化方法——Adam
#Adam是一种基于梯度的优化方法,
# 它结合了动量优化和RMSProp的优势,
# 可以在不同维度上有不同的学习率,
# 从而加速收敛并避免局部最优解。
#计算方法
#1.初始化学习率，权重参数，小常数ε，动量参数β1，RMSProp参数β2
#2.初始化梯度累计变量s=0，动量变量v=0
#3.从训练集中随机抽取多个样本，计算梯度g_t
#4.使用动量优化更新动量变量v_t：v_t=β1v_{t-1}+(1-β1)g_t
#5.使用RMSProp更新梯度累计变量s_t：s_t=β2s_{t-1}+(1-β2)g_t^2
#6.修正动量变量v_t：v_t=v_t/(1-β1^t)
#7.修正梯度累计变量s_t：s_t=s_t/(1-β2^t)
#8.使用学习率调整权重参数：w_t=w_{t-1}-η/sqrt(s_t+\varepsilon) *v_t

#api
#torch.optim.Adam(params, lr=0.001, betas=(0.9, 0.999), eps=1e-08, weight_decay=0, amsgrad=False)
#参数：
#params：待优化的参数组
#lr：学习率，默认值为0.001
#betas：动量参数，默认值为(0.9, 0.999)
#eps：小常数，默认值为1e-08
#weight_decay：权重衰减系数，默认值为0
#amsgrad：是否使用AMSGrad变体，默认值为False

#优缺点
#1. 优点：
# （1）自适应学习率：Adam根据梯度的平方进行自适应调整学习率,
# 可以在不同维度上有不同的学习率,
# 从而加速收敛。
# （2）平滑梯度：Adam通过对梯度的平方进行指数加权移动平均,
# 可以平滑梯度更新,
# 避免震荡和波动。
# （3）处理稀疏梯度：Adam对稀疏梯度有较好的处理能力,
# 可以在处理大规模数据集时保持较好的性能。
#2. 缺点：
# （1）学习率衰减问题：Adam的学习率衰减是指数级的,
# 可能会导致学习率过早衰减,
# 从而影响收敛速度。
# （2）对初始学习率敏感：Adam对初始学习率敏感,
# 不同的初始学习率可能会导致不同的收敛结果。

#Adam是一种常用的梯度下降优化方法,
# 它在许多深度学习任务中都取得了良好的效果,
# 被广泛应用于神经网络模型的训练中。

#示例
w=torch.tensor([5],dtype=torch.float32,requires_grad=True)
optimizer=torch.optim.Adam(params=[w],lr=0.1,eps=1e-8)
w_list=[]
for epoch in range(100):
    loss=0.5*w**2+20
    optimizer.zero_grad()
    loss.sum().backward()
    optimizer.step()
    w_list.append(w.item())
    print(w)
plt.figure(figsize=(10,8))
sns.lineplot(x=range(100),y=w_list)
plt.title('Adam Optimization')
plt.xlabel('Epoch')
plt.ylabel('w')
plt.show()
```

```python
#学习率优化
#学习率优化是指在训练神经网络模型时,通过调整学习率来加速模型的收敛速度和提高模型的性能。
#学习率优化的方法有很多,常见的有以下几种:
#1. 学习率衰减:在训练过程中逐渐降低学习率,可以使用指数衰减、阶梯衰减等方法。
#2. 学习率预热:在训练开始时使用较小的学习率,逐渐增加到预设的学习率,可以使用线性预热、指数预热等方法。
#3. 学习率循环:在训练过程中周期性地调整学习率,可以使用余弦循环、三角循环等方法。
#4. 学习率自适应:根据模型的训练情况自动调整学习率,可以使用ReduceLROnPlateau等方法。
#学习率优化可以帮助模型更快地收敛,避免陷入局部最优解,提高模型的性能。在实际应用中,可以根据具体的任务和数据集选择合适的学习率优化方法,并进行调试和调整以获得最佳的训练效果。

#为什么要进行学习率优化?
#1. 加速收敛:适当的学习率可以加速模型的收敛速度,使模型更快地达到最优解。
#2. 避免局部最优解:学习率优化可以帮助模型跳出局部最优解,找到全局最优解。
#3. 提高模型性能:通过调整学习率,可以提高模型的性能,使其在测试集上表现更好。

#学习率衰减方法——等间隔衰减
#等间隔衰减是指在训练过程中,每隔一定的迭代数就将学习率降低一定的比例,例如每隔10个epoch将学习率降低为原来的0.1倍。
#等间隔衰减的优点是简单易实现,可以在训练过程中逐渐降低学习率,加速模型的收敛速度。缺点是需要预先设定衰减的间隔和比例,可能需要进行调试和调整以获得最佳的训练效果。
#api
#torch.optim.lr_scheduler.StepLR(optimizer, step_size, gamma=0.1, last_epoch=-1)
#参数：
#optimizer：优化器对象
#step_size：学习率衰减的间隔,即每隔多少个epoch衰减一次学习率
#gamma：学习率衰减的比例,默认值为0.1
#last_epoch：上一个epoch的索引,默认值为-1

#示例
lr=0.1
max_epoch=200 #最大训练epoch数
iterations=10 #每个epoch迭代次数，即每个epoch进行10次梯度更新
y_true=torch.tensor([1],dtype=torch.float32)
x=torch.tensor([0],dtype=torch.float32,requires_grad=True)
w=torch.tensor([2],dtype=torch.float32,requires_grad=True)
optimizer=torch.optim.SGD(params=[w],lr=lr,momentum=0.9) #动量SGD优化器
optimizer_lr=torch.optim.lr_scheduler.StepLR(optimizer=optimizer,step_size=10,gamma=0.2) #每10个epoch衰减一次学习率,衰减为原来的0.2倍
lr_list,epoch_list=list(),list()
for epoch in range(max_epoch): #每个epoch进行10次梯度更新
    lr_list.append(optimizer_lr.get_last_lr()[0]) #获取当前epoch的学习率
    epoch_list.append(epoch)
    for _ in range(iterations):
        mse_loss=nn.MSELoss()
        loss=mse_loss(x,y_true)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    optimizer_lr.step() #一轮结束后更新学习率
plt.figure(figsize=(10,6))
sns.lineplot(x=epoch_list,y=lr_list)
plt.title('Learning Rate Decay - StepLR')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.show()

```
